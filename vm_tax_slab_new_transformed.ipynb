```python
import pandas as pd
```

```python
slab_new = pd.read_csv(r'D:\Vahan\Data\Jharkhand python\vm_tax_slab_new.csv')
```

```python
cpdf = pd.DataFrame({
    'class_type':[],
    'date_from':[],
    'date_to':[],
    'descr':[],
    'emp_cd':[],
    'op_dt':[], 
    'pur_cd':[], 
    'slab_code':[], 
    'state_cd':[], 
    'tax_mode':[],
    'vh_class':[]
})
```

```python
for ind , row in slab_new.iterrows():
    x=slab_new.at[ ind , 'tax_on_vch'].split(',')
    while '' in x: x.remove('')
    df = slab_new.loc[ind:ind,'class_type':'tax_mode']
    pdf = df
    for i in range(1, len(x)): pdf = pd.concat([pdf,df])
    if len(pdf)==len(x): 
        pdf['vh_class']=x
    cpdf= pd.concat([cpdf,pdf])
```

```python
cpdf.to_csv('D:/Vahan/Data/Jharkhand python/vm_tax_slab_new_transformed.csv',index=False)
```
